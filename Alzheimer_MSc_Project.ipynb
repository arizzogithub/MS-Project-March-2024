#### Data integrity and imputation strategy

This dataset encompasses highly sensitive and critical medical data for each of the 7,702 participants involved in this study. The data contains several special values representing categories such as 'Unknown,' 'Not available,' and 'Not applicable.' These values indicate instances where the data was either not collected, unavailable, or deemed irrelevant under certain circumstances. The presence of these special values across all participants presents a significant challenge for ensuring the robustness and accuracy of the machine learning models to be developed.

The original intent of this study was to analyze the data without resorting to any form of imputation, thus maintaining the highest level of data authenticity. In clinical practice, it is imperative that all relevant data be obtained and considered before making any diagnostic or therapeutic decisions, as the well-being and lives of patients are at stake. A clinician would typically require patients to retake or complete missing tests before proceeding with any conclusions, diagnoses or recommendations.

However, given the pervasive presence of these special values across the dataset, it has become clear that excluding participants with missing or ambiguous data would lead to a significant reduction in the sample size, potentially compromising the study's validity and generalizability. As a result, after careful consideration, the decision was made to employ data imputation techniques selectively. These techniques will be applied with the utmost caution, and only where absolutely necessary, with each attribute being evaluated on a case-by-case basis.

It is acknowledged that the use of data imputation carries the risk of introducing biases and potentially distorting the findings. However, this study is situated at the intersection of medical research and artificial intelligence, and its primary objective is to develop and optimize a multimodal machine learning model. The model aims to identify relationships between early markers and risk factors associated with Alzheimer’s disease (AD), and to classify participants based on their level of risk for developing AD in the future. In light of this objective, the application of data imputation is deemed an acceptable and necessary step to ensure the feasibility and integrity of the development process.

```python
# calculation based on new data.
# 1. calculating NACCBMI (body mass index) based on HEIGHT and WEIGHT
def calculate_bmi(row):
    if row['HEIGHT'] > 0 and row['WEIGHT'] > 0:  # check that the values are positive
        return (row['WEIGHT'] * 703) / (row['HEIGHT'] ** 2)
    else:
        return pd.NA  # if the data is incorrect, return NA

df['NACCBMI'] = df.apply(calculate_bmi, axis=1)

# 2. calculating NACCNE4S based on NACCAPOE
def calculate_naccne4s(naccapoe):
    if naccapoe == 2:  # e3,e4
        return 1
    elif naccapoe == 4:  # e4,e4
        return 2
    elif naccapoe == 5:  # e4,e2
        return 1
    else:
        return 0  # for all other cases (e3,e3; e3,e2; e2,e2)

df['NACCNE4S'] = df['NACCAPOE'].apply(calculate_naccne4s)

# checking the results
print(df[['NACCBMI', 'NACCNE4S']].head())
```

#### Define 'Main Risk Group' attribute

The datasets lack the primary attribute required for the multiclass classification task. This attribute will be generated through feature creation to ensure the dataset meets the necessary requirements for accurate predictive modeling.

**Defining 6 Main Risk Groups:**

* High Risk 
* Moderate Risk
* Low Risk
* Primary AD diagnosis
* Contributing AD diagnosis
* Non-AD or Uncertain Status
 
 
In the provided code, a distinct logic is implemented for participants diagnosed with Alzheimer's disease (AD). These individuals are assigned a risk score of zero, effectively excluding them from further risk assessment. This approach ensures that participants with an AD diagnosis are categorized into specific risk groups—either as having a "Primary AD diagnosis" or a "Contributing AD diagnosis." By doing so, the code acknowledges the significance of an AD diagnosis in determining cognitive impairment, rendering additional risk factor analysis unnecessary for these individuals.

The "Non-AD or Uncertain Status" group encompasses all participants for whom Alzheimer's disease (AD) is either not confirmed or whose AD status is unclear. These participants are excluded from further risk assessment and are assigned a risk score of zero.

**Distribution of Scores Across Risk Groups:**

* Low Risk Group: 0–16 points (approximately 0%–33% of the maximum possible score)
* Moderate Risk Group: 17–33 points (approximately 34%–66% of the maximum possible score)
* High Risk Group: 34–50 points (approximately 67%–100% of the maximum possible score)

```python
def detailed_risk_classification_with_scores(df_train):
    risk_groups = []
    risk_scores = []
    main_risk_groups = []
    level_of_risks = []

    for index, row in df_train.iterrows():
        group = []
        risk_score = 0  # initial risk score
    
        # check for AD diagnosis and categorize
        if row['NACCALZD'] == 1:
            if row['NACCALZP'] == 1:
                group.append('Primary AD diagnosis')
                main_risk_groups.append('Primary AD diagnosis')
                level_of_risks.append('Primary AD diagnosis')
                risk_scores.append(0)  # exclude from scoring
            elif row['NACCALZP'] == 2:
                group.append('Contributing AD diagnosis')
                main_risk_groups.append('Contributing AD diagnosis')
                level_of_risks.append('Contributing AD diagnosis')
                risk_scores.append(0)  # exclude from scoring
            else:
                group.append('Non-AD or Uncertain Status')
                main_risk_groups.append('Non-AD or Uncertain Status')
                level_of_risks.append('Non-AD or Uncertain Status')
                risk_scores.append(0)  # exclude from scoring

            
            # even if continue is executed, we need to ensure the lists are updated
            risk_groups.append(', '.join(group))
            continue  # move to the next participant

        # gender consideration
        if row['SEX'] == 2:  # female
            risk_score += 2  # moderate risk for females
            group.append('Female')
        elif row['SEX'] == 1:  # male
            risk_score += 1  # low risk for males
            group.append('Male')

        # age
        if row['NACCMRIA'] > 75:
            group.append('Age over 75')
            risk_score += 3  # high risk
        elif 65 <= row['NACCMRIA'] <= 75:
            group.append('Age 65-75')
            risk_score += 2  # moderate risk
        elif row['NACCMRIA'] < 65:
            group.append('Age under 65')
            risk_score += 1  # low risk

        # genetic predisposition
        if row['NACCNE4S'] == 2:
            group.append('Two copies of APOE ε4')
            risk_score += 3  # high risk
        elif row['NACCNE4S'] == 1:
            group.append('One copy of APOE ε4')
            risk_score += 2  # moderate risk
        else:
            group.append('No APOE ε4')
            risk_score += 1  # low risk

        # specific APOE genotype
        if row['NACCAPOE'] == 4:
            group.append('APOE genotype e4,e4')
            risk_score += 3  # high risk
        elif row['NACCAPOE'] in [2, 5]:
            group.append('APOE genotype e3,e4 or e4,e2')
            risk_score += 2  # moderate risk
        elif row['NACCAPOE'] in [1, 3, 6]:
            group.append('APOE genotype e3,e3 or e3,e2 or e2,e2')
            risk_score += 1  # low risk

        # cognitive status
        if row['NACCUDSD'] == 3:
            group.append('MCI (moderate risk)')
            risk_score += 2  # moderate risk
        elif row['NACCUDSD'] == 2:
            group.append('Mild cognitive impairment')
            risk_score += 1  # low risk
        elif row['NACCUDSD'] == 1:
            group.append('Normal cognitive function')
            risk_score += 0  # low risk

        # family history (parents)
        if row['NACCMOM'] == 1 and row['NACCDAD'] == 1:
            group.append('Cognitive impairment in both parents')
            risk_score += 3  # high risk
        elif row['NACCMOM'] == 1 or row['NACCDAD'] == 1:
            group.append('Cognitive impairment in one parent')
            risk_score += 2  # moderate risk
        else:
            group.append('No cognitive impairment in parents')
            risk_score += 1  # low risk

        # cognitive test results
        if row['NACCMMSE'] < 24:
            group.append('MMSE < 24 (low score)')
            risk_score += 3  # high risk
        elif 24 <= row['NACCMMSE'] < 27:
            group.append('MMSE 24-26 (moderate score)')
            risk_score += 2  # moderate risk
        else:
            group.append('MMSE ≥ 27 (normal score)')
            risk_score += 1  # low risk

        if row['NACCMOCA'] < 26:
            group.append('MoCA < 26 (low score)')
            risk_score += 3  # high risk
        elif 26 <= row['NACCMOCA'] < 28:
            group.append('MoCA 26-27 (moderate score)')
            risk_score += 2  # moderate risk
        else:
            group.append('MoCA ≥ 28 (normal score)')
            risk_score += 1  # low risk

        if row['CDRSUM'] >= 4.0:
            group.append('CDRSUM ≥ 4.0 (significant impairment)')
            risk_score += 3  # high risk
        elif 2.0 <= row['CDRSUM'] < 4.0:
            group.append('CDRSUM 2.0-3.9 (moderate impairment)')
            risk_score += 2  # moderate risk
        else:
            group.append('CDRSUM < 2.0 (minor impairment)')
            risk_score += 1  # low risk

        if row['CDRGLOB'] >= 2.0:
            group.append('CDRGLOB ≥ 2.0 (moderate or severe dementia)')
            risk_score += 3  # high risk
        elif row['CDRGLOB'] == 1.0:
            group.append('CDRGLOB = 1.0 (mild dementia)')
            risk_score += 2  # moderate risk
        else:
            group.append('CDRGLOB < 1.0 (normal cognition or MCI)')
            risk_score += 1  # low risk

        # education
        if row['EDUC'] < 9:
            group.append('Low level of education')
            risk_score += 3  # high risk
        elif 9 <= row['EDUC'] <= 12:
            group.append('Medium level of education')
            risk_score += 2  # moderate risk
        else:
            group.append('High level of education')
            risk_score += 1  # low risk

        # chronic diseases
        if row['DIABETES'] == 1:
            group.append('Diabetes')
            risk_score += 2  # moderate risk
        if row['HYPERTEN'] == 1:
            group.append('Hypertension')
            risk_score += 2  # moderate risk
        if row['THYROID'] == 1:
            group.append('Thyroid disease')
            risk_score += 1  # low risk

        # alcohol abuse and depression
        if row['ALCOHOL'] == 1:
            group.append('Alcohol dependence')
            risk_score += 2  # moderate risk
        if row['DEP2YRS'] == 1 or row['DEPOTHR'] == 1:
            group.append('Depression')
            risk_score += 2  # moderate risk

        # smoking impact
        if row['SMOKYRS'] > 20:
            group.append('Smoking for more than 20 years')
            risk_score += 2  # moderate risk
        elif 1 <= row['SMOKYRS'] <= 20:
            group.append('Smoking for up to 20 years')
            risk_score += 1  # low risk
        else:
            group.append('Non-smoker')

        # body mass index (BMI)
        if row['NACCBMI'] >= 30:
            group.append('Obesity')
            risk_score += 2  # moderate risk
        elif row['NACCBMI'] < 18.5:
            group.append('Underweight')
            risk_score += 2  # moderate risk
        else:
            group.append('Normal weight')
            risk_score += 1  # low risk

        # hearing and vision impairments
        if row['VISION'] == 1:
            group.append('Vision impairment')
            risk_score += 1  # low risk
        if row['HEARING'] == 1:
            group.append('Hearing impairment')
            risk_score += 1  # low risk

        # socio-demographic factors
        if row['MARISTAT'] in [2, 3, 4, 5]:
            group.append('Without a partner')
            risk_score += 1  # low risk
        if row['NACCLIVS'] == 1:
            group.append('Lives alone')
            risk_score += 2  # moderate risk
        if row['INDEPEND'] >= 2:
            group.append('Needs help in daily life')
            risk_score += 2  # moderate risk

        # assign main risk group
        if risk_score <= 16:
            main_risk_group = 'Low Risk'
        elif 17 <= risk_score <= 33:
            main_risk_group = 'Moderate Risk'
        else:
            main_risk_group = 'High Risk'
        main_risk_groups.append(main_risk_group)

        # add group and risk score to the lists
        risk_groups.append(', '.join(group))
        risk_scores.append(risk_score)

    df_train['Detailed_Risk_Group'] = risk_groups
    df_train['Risk_Score'] = risk_scores
    df_train['Main_Risk_Group'] = main_risk_groups
   
    return df_train

# apply the function to your DataFrame
df_train = detailed_risk_classification_with_scores(df_train)

# save the results to a CSV file
df_train.to_csv('M_1_df_train_risk_classification_results.csv', index=False)
```

#### SMOTE for each class (Over-sampling)

```python
def oversample_to_first_class(df_train, target_column):
    """
    Performs oversampling for classes with fewer participants to the level of the first class.

    Parameters:
    - df_train: DataFrame with the data.
    - target_column: the name of the column with class labels.

    Returns:
    - dataset_over: DataFrame with oversampled data.
    """
    # split the data by class
    class_counts = df_train[target_column].value_counts()
    first_class = class_counts.idxmax()
    count_first_class = class_counts.max()

    # perform oversampling for each class
    oversampled_classes = []
    for class_label in class_counts.index:
        class_subset = df_train[df_train[target_column] == class_label]
        if class_label != first_class:
            class_subset = class_subset.sample(count_first_class, replace=True)
        oversampled_classes.append(class_subset)

    # combine the oversampled data
    dataset_over = pd.concat(oversampled_classes, axis=0)

    return dataset_over

# example usage:
dataset_over = oversample_to_first_class(df_train, 'Main_Risk_Group')

# check the results
print('Random over-sampling:')
print(dataset_over['Main_Risk_Group'].value_counts())

# visualization
dataset_over['Main_Risk_Group'].value_counts().plot(kind='bar', title='Count (Main_Risk_Group)');

df_train = dataset_over
```

Even with a fixed random seed for reproducibility, minor fluctuations in performance can still occur due to these inherent random processes.

However, these variations do not significantly impact predictions on test data, as models tuned with regularization and cross-validation exhibit strong generalization ability. The small variability in test results reflects the balance between the model's adaptation to training data and its ability to avoid overfitting, remaining robust when applied to new data.

#### Build and train models

Hyperparameters play a vital role in optimizing machine learning models, influencing their learning process and overall performance. Unlike parameters, which are derived from the data, hyperparameters are predefined settings adjusted before training. They include aspects such as the number of trees in a Random Forest or the learning rate in Gradient Boosting. Effective tuning of hyperparameters can enhance model accuracy, efficiency, and generalization. 

In the **RandomForestClassifier**, key hyperparameters include:

- **bootstrap=True**: Enables bootstrapping, where each tree is trained on a random subset of the data with replacement. 
- **criterion='gini'**: Measures the quality of splits based on Gini impurity, which quantifies label distribution purity.
- **max_depth=None**: Allows trees to grow until leaves are pure or contain fewer samples than `min_samples_split`, with no depth limit.
- **max_features='log2'**: Considers the logarithm base 2 of the total number of features for each split, introducing randomness to reduce overfitting.
- **min_samples_leaf=1**: Requires at least one sample per leaf, enabling very fine splits and potential overfitting.
- **min_samples_split=2**: Allows internal nodes to split with as few as two samples, creating complex tree structures.
- **n_estimators=100**: Defines the number of trees, balancing computational efficiency and accuracy.
- **random_state=42**: Ensures reproducibility by fixing the random number generator seed.

In the **GradientBoostingClassifier**, key hyperparameters include:

- **learning_rate=0.2**: Controls the contribution of each tree to the overall model, balancing between training speed and predictive accuracy.
- **max_depth=5**: Limits the maximum depth of each individual tree, preventing overfitting while maintaining model complexity.
- **min_samples_leaf=4**: Requires at least four samples in a leaf node, reducing overfitting by preventing overly fine splits.
- **min_samples_split=5**: Ensures that internal nodes have at least five samples to allow a split, promoting a more generalized model.
- **n_estimators=300**: Defines the number of boosting stages (trees) to be used, balancing accuracy and computational cost.
- **subsample=0.8**: Uses 80% of the data for each tree, injecting randomness to enhance model robustness and reduce overfitting.
- **random_state=42**: Ensures reproducibility by fixing the seed for the random number generator, allowing consistent results.

In the **MLPClassifier**, key hyperparameters include:

- **hidden_layer_sizes=(100, 50)**: Defines a neural network with two hidden layers, consisting of 100 and 50 neurons respectively, shaping the model's capacity.
- **activation='tanh'**: Uses the hyperbolic tangent function for activation, mapping outputs to the range -1 to 1.
- **solver='adam'**: Applies the Adam optimizer for efficient gradient-based optimization with adaptive learning rates.
- **alpha=0.01**: Sets regularization strength to 0.01, penalizing large weights to reduce overfitting.
- **learning_rate='adaptive'**: Adjusts the learning rate based on performance, enhancing convergence.
- **learning_rate_init=0.001**: Starts with an initial learning rate of 0.001, controlling the step size during training.
- **momentum=0.95**: Utilizes momentum to accelerate training and smooth weight updates.
- **n_iter_no_change=10**: Halts training if no improvement is observed in the validation score over 10 iterations.
- **random_state=42**: Ensures reproducibility by setting a fixed seed for random number generation.
- **max_iter=500**: Limits the training iterations to 500, balancing training time and convergence.nvergence.


To rigorously assess the performance of models, **5-fold cross-validation** was employed. This method involves partitioning the training data into five subsets, where each subset is used in turn as a validation set while the remaining four subsets serve as the training set. By iterating this process, a range of accuracy scores that reflect the model’s performance across different data splits is generated. This approach not only provides a more reliable estimate of the model's accuracy but also helps in mitigating overfitting by ensuring the model is evaluated on diverse subsets of the training data. This cross-validation procedure is critical for comparing the Random Forest model's efficacy against other classifiers and for validating its robustness in generalizing to unseen data.

#### Metrics: Classification report

__Classification Report__ is a text report showing the main classification metrics:

* __Precision__: The ratio of correctly predicted positive observations to the total predicted positive observations. Formula: 
     Precision = TP / (TP + FP)
    
* __Recall__: The ratio of correctly predicted positive observations to all observations in the actual class. Formula:
    Recall = TP / (TP + FN)

* __F1__: The harmonic mean of Precision and Recall. Formula:
    F1 Score = 2*(Recall * Precision) / (Recall + Precision)
    
* __Support__: The number of actual occurrences of each class in the specified dataset.

* __Accuracy__: The ratio of correctly predicted observations to the total observations. Formula:
    Accuracy = TP+TN / TP + FP + FN + TN
    
* __Macro-average__: The average of metrics calculated for each class, treating all classes equally regardless of their support.
  
* __Weighted average__:The average of metrics calculated for each class, weighted by the support (number of true instances) for each class.

#### Metrics: ROC-AUC

ROC-AUC (Receiver Operating Characteristic - Area Under the Curve) quantifies a model’s ability to distinguish between classes, offering a summary of performance across all thresholds.

```python
# function to plot ROC curve on a specific axis and calculate average ROC AUC
def plot_roc_curve(ax, y_val_bin, y_pred_proba, n_classes, model_name):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_val_bin[:, i], y_pred_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label=f'Class {i} (area = {roc_auc[i]:.2f})')

    ax.plot([0, 1], [0, 1], 'k--', lw=2)  # random line (baseline)
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC-AUC - {model_name}')
    ax.legend(loc="lower right", fontsize='small')

    # return the average ROC AUC across all classes
    return sum(roc_auc.values()) / n_classes

# binarizing the class labels
y_val_bin = label_binarize(y_val, classes=[0, 1, 2, 3, 4, 5])  # specify your classes
n_classes = y_val_bin.shape[1]

# create figure and axes for 3 graphs in a row
fig, axs = plt.subplots(1, 3, figsize=(18, 5))  # 1 row, 3 columns

#  Random Forest
y_pred_proba_rf = RF1.predict_proba(X_val)
roc_auc_rf = plot_roc_curve(axs[0], y_val_bin, y_pred_proba_rf, n_classes, "Random Forest")

# Gradient Boosting
y_pred_proba_gb = GB1.predict_proba(X_val)
roc_auc_gb = plot_roc_curve(axs[1], y_val_bin, y_pred_proba_gb, n_classes, "Gradient Boosting")

# Multi-layer Perceptron
y_pred_proba_mlp = MLP1.predict_proba(X_val)
roc_auc_mlp = plot_roc_curve(axs[2], y_val_bin, y_pred_proba_mlp, n_classes, "Multi-Layer Perceptron")

# display all graphs
plt.tight_layout()
plt.show()

# print ROC-AUC results for each model
print(f"Random Forest ROC-AUC: {roc_auc_rf:.2f}")
print(f"Gradient Boosting ROC-AUC: {roc_auc_gb:.2f}")
print(f"Multi-Layer Perceptron ROC-AUC: {roc_auc_mlp:.2f}")
```

#### 'Confidence score' and visualisation for Test dataset predictions 

Confidence scores for predictions on unseen data provide a measure of the certainty or probability associated with each prediction. They are essential for understanding the model's confidence in its predictions and identifying instances where the model may be uncertain or less reliable. High confidence scores suggest that the model is sure about its predictions, while lower scores indicate potential uncertainty.

The accuracy of confidence scores depends on the calibration of the model. A well-calibrated model ensures that its confidence scores accurately reflect the true likelihood of correct predictions. However, if the model is poorly calibrated, confidence scores may not reliably represent the true probability of correctness.

```python
# function to print predictions and confidence scores
def model_predictions(model, model_name, X_Test_1):
    # class predictions
    test_predictions = model.predict(X_Test_1)

    # probability predictions (confidence scores)
    test_probabilities = model.predict_proba(X_Test_1)

    # calculate the confidence score for each prediction
    confidence_scores = np.max(test_probabilities, axis=1)

    # bar chart for the number of predictions per class and average confidence score 
    class_counts = np.unique(test_predictions, return_counts=True)
    avg_confidence_per_class = [np.mean(confidence_scores[test_predictions == class_label]) for class_label in np.unique(test_predictions)]

    # two charts side by side: number of predictions per class and average confidence score
    fig, axes = plt.subplots(1, 2, figsize=(9, 3))

    # number of predictions per class
    sns.barplot(x=class_counts[0], y=class_counts[1], ax=axes[0], palette='Blues_d')
    axes[0].set_title('Number of Predictions per Class', fontsize=14)
    axes[0].set_xlabel('Classes', fontsize=10)
    axes[0].set_ylabel('Count', fontsize=10)

    # add the count values on top of each bar
    for i in range(len(class_counts[0])):
        axes[0].text(i, class_counts[1][i] + 0.5, str(class_counts[1][i]), ha='center', fontsize=7)

    # average confidence score by class
    sns.barplot(x=np.unique(test_predictions), y=avg_confidence_per_class, ax=axes[1], palette='Greens_d')
    axes[1].set_title('Average Confidence Score by Class', fontsize=14)
    axes[1].set_xlabel('Classes', fontsize=10)
    axes[1].set_ylabel('Average Confidence Score', fontsize=10)

    # add the average confidence score values on top of each bar
    for i in range(len(avg_confidence_per_class)):
        axes[1].text(i, avg_confidence_per_class[i] + 0.01, f'{avg_confidence_per_class[i]:.2f}', ha='center', fontsize=7)

    plt.tight_layout()
    plt.show()

    # density plot for confidence scores 
    plt.figure(figsize=(8, 5))
    for class_label in np.unique(test_predictions):
        sns.kdeplot(confidence_scores[test_predictions == class_label], label=f'Class {class_label}', fill=True)
    plt.title('Density Plot of Confidence Scores by Classes', fontsize=14)
    plt.xlabel('Confidence Score', fontsize=10)
    plt.ylabel('Density', fontsize=10)
    plt.legend(title='Classes')
    plt.grid(True)
    plt.show()

    # print confidence scores for the first 5 samples of the test set
    print(f"\n--- {model_name} Model Predictions ---")
    for i, score in enumerate(confidence_scores[:5]):
        print(f"Sample {i+1}: predicted class = {test_predictions[i]}, confidence score = {score:.2f}")

    # calculate the average confidence score across the entire test set
    average_confidence = np.mean(confidence_scores)
    print(f"\n{model_name} Average confidence score on the test set: {average_confidence:.2f}")

# predictions and confidence scores for each model 
print("\nRandom Forest:")
model_predictions(RF1, "Random Forest", X_Test_1)

print("\nGradient Boosting:")
model_predictions(GB1, "Gradient Boosting", X_Test_1)

print("\nMLP:")
model_predictions(MLP1, "MLP", X_Test_1)
```

As can be seen, the average confidence scores for predictions on the test set reveal notable differences among the models. The Random Forest classifier achieved an average confidence score of 0.92, indicating a high level of certainty in its predictions. The Gradient Boosting model demonstrated a slightly higher average confidence score of 0.99, suggesting an even greater degree of predictive certainty. In contrast, the Multi-layer Perceptron (MLP) classifier exhibited the highest average confidence score at 0.99, reflecting an exceptionally high level of confidence in its predictions. ### Integration of calculations between the 2nd and 4th modalities

Let's examine the second and fourth modalities.

The second modality data includes the attributes **WHITEVOL (Total brain white matter volume, cc)** and **WMHVOL (Total brain white matter hyperintensity volume, cc)**.

The fourth modality data includes the attributes **Lesion_Vol (Total brain lesions volume, ml)** and **Num_of_Lesions (Number of brain lesions)**.

(Please note that cubic centimeters (cc) and milliliters (ml) are equivalent units of volume, where 1 cc = 1 cm³ = 1 ml).

**WMHVOL** is a key indicator used to identify pathological changes in the brain’s white matter. Hyperintensities, visible on MRI as bright spots, are often associated with small vessel damage in the brain and, in the context of Alzheimer’s disease, may indicate vascular disruptions that impact cognitive function.

**Lesion_Vol** is a broader measure that captures all types of white matter damage, not just hyperintensities. This is crucial for understanding the full spectrum of changes in white matter, as other forms of damage can also be linked to neurodegenerative processes typical of Alzheimer’s disease.

While WMHVOL and Lesion_Vol may overlap (with hyperintensities being a form of lesion), they are not identical. WMHVOL reflects a specific type of damage, while Lesion_Vol encompasses a wider range of changes. Comparing these two measures provides a more comprehensive picture of white matter integrity, allowing for greater accuracy in analysis and deeper clinical insights. This can improve predictions of Alzheimer's disease progression and enable earlier interventions.

The next step involves classifying brain white matter status based on these two key metrics: WMHVOL and Lesion_Vol. These parameters will be used to assess the degree of neurodegenerative changes associated with Alzheimer’s disease.

Initially, the lesion and hyperintensity percentages relative to the total white matter volume will be calculated, resulting in two new parameters: **'Lesion_Percent' and 'WMH_Percent'**. Participants will then be grouped based on the hyperintensities, and the level and number of lesions **(Num_of_Lesions)**, producing two additional parameters: **'Lesion_Percent_Category' and 'WMH_Percent_Category'**. This will help identify specific vascular disruptions (hyperintensities) and broader white matter damage, providing deeper insights into disease progression. Higher percentages may serve as markers of significant deterioration, while lower percentages could indicate earlier stages of the disease.

Following this, a **correlation** will be performed between the percentage of white matter lesions (Lesion_Percent) and the percentage of white matter hyperintensities (WMH_Percent) in this study cohort.

The classification of lesion percentages is based on **four groups**, with an additional distinction made by the number of lesions. This approach allows for a more detailed understanding of the severity of brain lesions.

**Minor Lesion Percentage (< 2%):**

This group is generally considered normal in younger individuals or those without notable vascular risk factors. It can also be observed in the early stages of cognitive decline. In this range, further differentiation is made between cases with a low number of lesions (≤ 10) and a high number of lesions (> 10), as a higher lesion count might indicate increased vulnerability.

**Moderate Lesion Percentage (2 <= % <= 5):**

This range is typically associated with mild vascular alterations or early cognitive impairment. It is more frequently observed in middle-aged and older individuals, especially in the presence of risk factors such as hypertension or diabetes. Here, again, lesions are classified based on whether the number of lesions is low (≤ 10) or high (> 10), reflecting the potential influence of lesion quantity on the severity of cognitive impairment.

**Moderate to Severe Lesion Percentage (5 < % <= 7):**

 early-stage Alzheimer's disease, or significant vascular alterations. The number of lesions plays a role in determining the impact on cognitive function, with both low (≤ 10) and high (> 10) lesion counts further refining the classification.

**Severe Lesion Percentage (> 7%):**

A high lesion percentage in this group is commonly observed in advanced cognitive impairment, including late-stage Alzheimer's disease. Pronounced vascular changes, such as chronic cerebral ischemia, are often present. Lesion severity is additionally categorized by whether the number of lesions is low (≤ 10) or high (> 10), with larger lesion counts suggesting more extensive brain damage.

```python
def classify_lesion_percentage(percent, count):
    if percent < 2 and count <= 10:
        return 'Minor lesions % (low number of lesions)'
    elif percent < 2 and count > 10:
        return 'Minor lesions % (high number of lesions)'
    elif 2 <= percent <= 5 and count <= 10:
        return 'Moderate lesions (low number of lesions)'
    elif 2 <= percent <= 5 and count > 10:
        return 'Moderate lesions (high number of lesions)'
    elif 5 < percent <= 7 and count <= 10:
        return 'Moderate to Severe lesions (low number of lesions)'
    elif 5 < percent <= 7 and count > 10:
        return 'Moderate to Severe lesions (high number of lesions)'
    elif percent > 7 and count <= 10:
        return 'Severe lesions (low number of lesions)'
    else:
        return 'Severe lesions (high number of lesions)'

def classify_wmh_percentage(percent):
    if percent < 2:
        return 'Minor WMH %'
    elif 2 <= percent <= 5:
        return 'Moderate WMH %'
    elif 5 < percent <= 7:
        return 'Moderate to Severe WMH %'
    else:
        return 'Severe WMH %'

def classify_lesion_and_wmh_percentages(df):
    """
    adds columns to the dataframe classifying the percentages of white matter lesions and white matter hyperintensities.

    :param df: dataframe with columns 'Lesion_Vol', 'WMHVOL', and 'WHITEVOL'
    :return: dataframe with additional columns 'Lesion_Percent_Category' and 'WMH_Percent_Category'
    """
    # calculate lesion percentage
    df['Lesion_Percent'] = (df['Lesion_Vol'] / df['WHITEVOL']) * 100

    # calculate wmh percentage
    df['WMH_Percent'] = (df['WMHVOL'] / df['WHITEVOL']) * 100

    # apply classifications
    df['Lesion_Percent_Category'] = df.apply(lambda row: classify_lesion_percentage(row['Lesion_Percent'], row['Lesion_Vol']), axis=1)
    df['WMH_Percent_Category'] = df['WMH_Percent'].apply(classify_wmh_percentage)

    return df

# classify data
df = classify_lesion_and_wmh_percentages(df)

# create dictionaries to map text labels to numerical values
lesion_mapping = {
    'Minor lesions % (low number of lesions)': 0,
    'Minor lesions % (high number of lesions)': 1,
    'Moderate lesions (low number of lesions)': 2,
    'Moderate lesions (high number of lesions)': 3,
    'Moderate to Severe lesions (low number of lesions)': 4,
    'Moderate to Severe lesions (high number of lesions)': 5,
    'Severe lesions (low number of lesions)': 6,
    'Severe lesions (high number of lesions)': 7
}

wmh_mapping = {
    'Minor WMH %': 0,
    'Moderate WMH %': 1,
    'Moderate to Severe WMH %': 2,
    'Severe WMH %': 3
}

# map text labels to numerical values for both categories
df['Lesion_Percent_Category'] = df['Lesion_Percent_Category'].map(lesion_mapping).astype('category')
df['WMH_Percent_Category'] = df['WMH_Percent_Category'].map(wmh_mapping).astype('category')

# save the results to a csv file
df.to_csv('fusion_risk_classification_%_results.csv', index=False)

# display the updated dataframe with classifications
df.head()
```

#### Risk Group Classification and Label Mapping

This code maps combinations of diagnostic risk groups, white matter hyperintensity categories, and lesion categories to specific risk group numbers and descriptions. It first converts categorical labels into numeric values using predefined dictionaries, then generates all possible combinations of these categories. Each combination is assigned a unique risk group number and a corresponding description. There are 192 possible combinations or risk groups, and this is saved in a separate CSV file. The code then applies these mappings to a dataset, creating two new attributes: one for the risk group number (Risk_Group) and another for the description (Description). #### SMOTE for each class (Over-sampling)

```python
def oversample_to_first_class(df_train, target_column):
    """
    Performs oversampling for classes with fewer participants to the level of the first class.

    Parameters:
    - df_train: DataFrame with the data.
    - target_column: the name of the column with class labels.

    Returns:
    - dataset_over: DataFrame with oversampled data.
    """
    # split the data by class
    class_counts = df_train[target_column].value_counts()
    first_class = class_counts.idxmax()
    count_first_class = class_counts.max()

    # perform oversampling for each class
    oversampled_classes = []
    for class_label in class_counts.index:
        class_subset = df_train[df_train[target_column] == class_label]
        if class_label != first_class:
            class_subset = class_subset.sample(count_first_class, replace=True)
        oversampled_classes.append(class_subset)

    # combine the oversampled data
    dataset_over = pd.concat(oversampled_classes, axis=0)

    return dataset_over

# usage
dataset_over = oversample_to_first_class(df_train, 'Risk_Group')

# check the results
#print('Random over-sampling:')
#print(dataset_over['Risk_Group'].value_counts())

# visualization
#dataset_over['Risk_Group'].value_counts().plot(kind='bar', title='Count (Risk_Group)');

df_train = dataset_over
```

The high performance across both phases suggests that the model is well-suited for early identification of at-risk individuals, supporting its potential application in preventive AD strategies.
